# What is ```Matplotlib?```
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

````python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o',
         ms=15,
         mfc='r',#marker face color
         mec='b',#marker edge color
         color='black',
         ls="--",#line space 
         alpha=0.7,#blurness
         label='lineplot')
plt.savefig('plot_attributes.png')
plt.show()
````

````python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o',
         ms=15,
         mfc='r',#marker face color
         mec='b',#marker edge color
         color='black',
         ls="--",#line space 
         alpha=0.7,#blurness
         label='lineplot',
         )
plt.title('Matplotlib_Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.savefig('plot_methods.png')
plt.show()
````

# subplot

````python
x=np.linspace(0,10,100)
plt.Figure(figsize=(10, 6))
plt.subplot(2,2,1)# it will create multiple plots of size 2
plt.plot(x,np.sin(x))
plt.title("Sin Wave")
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("Cos-Wave")
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.title("Tan-Wave")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("Log-Wave")
plt.tight_layout()# this prevent from overlapping
plt.savefig('Subplot.png')
plt.show()
````
