# What is ```Seaborn?```
Seaborn is a powerful Python data visualization library built on top of matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy  as np
````

````python
df=sns.load_dataset("tips")
print(df)
print(df.head())
````

### Basic plots
```1. Scatter Plot```

````python
sns.scatterplot(data=df,x="total_bill",y="tip",hue="sex")
plt.show()
````

```2. Line Plot```

````python
sns.lineplot(data=df,x="total_bill",y="tip",hue="smoker",)
plt.show()
````

```3. Bar Plot```


````python
sns.barplot(data=df, x="day", y="total_bill")
sns.set_style("darkgrid")
plt.show()
````

``` 4. Histogram```

````python
sns.histplot(data=df, x="total_bill", kde=True)
plt.show()
````

``` 5. box plot

``` 5. box plot```
it means spread of data

````python
sns.boxplot(data=df, x="day", y="total_bill",hue="smoker",)
plt.show()
````

```6. Violin plot``` it means distribution of data

````python
sns.violinplot(data=df, x="day", y="total_bill")
plt.show()
````

````python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    #style="smoker",
   # palette="Set2"
)
plt.title("Tip vs Total Bill")
plt.show()
````

````python
sns.barplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    color="red",
    #size="size",
    #style="smoker",
    palette="Set2"
)
plt.title("Tip vs Total Bill")
plt.show()
````

```3. Intermediate Plots```

1. Count Plot

````python
sns.countplot(data=df, x="day")
plt.show()
````

```2. Heatmap (correlation matrix)```

````python

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
````

```3. Pair Plot```

````python
sns.pairplot(df)
plt.show()
````

```4. Joint Plot```

````python
sns.jointplot(data=df, x="total_bill", y="tip", kind="hex")
plt.show()
````

```# 5. Advanced Usage```
1. FacetGrid(multiple plots for subgroups)

````python
g = sns.FacetGrid(df, col="sex", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
````

```2. Catplot(categorical plot)```

````python
sns.catplot(data=df, x="day", y="total_bill", kind="violin", hue="sex")
plt.show()
````

```3. Swarm plot```

````python
sns.swarmplot(data=df, x="day", y="tip", hue="sex")
plt.show()
````

````python
#relplot
sns.relplot(data=df, x="total_bill", y="tip", hue="sex",kind="line",)

plt.show()
````

#### calculating outliers

````python
Q1 = df['total_bill'].quantile(0.25)
Q3 = df['total_bill'].quantile(0.75)
IQR = Q3 - Q1

# Outlier range
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['total_bill'] < lower) | (df['total_bill'] > upper)]
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Custom plot
sns.scatterplot(data=df, x="total_bill", y="tip", hue="smoker", style="smoker", size="size")
plt.title("Total Bill vs Tip")               # Custom title
plt.xlabel("Bill Amount ($)")                # Custom X-axis label
plt.ylabel("Tip Amount ($)")                 # Custom Y-axis label
plt.legend(title="Customer Info")            # Custom legend
plt.grid(True)                               # Add gridlines
plt.show()
````
